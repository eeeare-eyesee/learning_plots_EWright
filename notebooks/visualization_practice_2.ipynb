```python
pylot.figure(figsize=(10,6))
pylot.scatter(spotify_df["Streams Last 30 Days (Millions)"], 
              spotify_df["Total Streams (Millions)"],
              color = "#08cc15",
              alpha = 0.8,
              s=10)
pylot.xlabel("Streams this month")
pylot.ylabel("Total Streams")

pylot.title("No apparent correlation between monthly listens and total streams")
ax = pylot.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

pylot.show()
```

```python
pylot.figure(figsize=(10,6))

featured_genre = "Pop"
colors = ["#08cc15" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

pylot.scatter(spotify_df["Streams Last 30 Days (Millions)"], 
              spotify_df["Total Streams (Millions)"],
              color =colors,
              alpha = 0.8,
              s=10)
pylot.xlabel("Streams this month")
pylot.ylabel("Total Streams")

pylot.title("No apparent correlation between monthly listens and total streams")
ax = pylot.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do pop songs achieve higher monthly listens or total streams?", fontsize = 18, pad=45)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

pylot.show()
```

```python
yearly_avg=spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg

```

```python
pylot.figure(figsize = (10,6))
pylot.plot(
    yearly_avg.index,
    yearly_avg.values,
    marker="+",
    linestyle="--",
    color= "#20a32b",
    linewidth = 0.5)

pylot.xlabel("Release Year")
pylot.ylabel("Average Total Streams")
pylot.title("Did listeners stream less in 2022", fontsize = 18,pad=20)

ax=pylot.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color = "#063b0a",
    linewidth= 0.5,
    alpha  = 0.7
)

ax.text(
    2022+0.05,
    3000,
    "Dip in total streams",
    color = "#063b08",
    fontsize=8
)


pylot.ylim(0,3500)
pylot.show()
```

```python
category = "Platform Type"
values="Total Streams (Millions)"
platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
pylot.figure(figsize=(8,8))
pylot.pie(
    platform_totals,
    labels=platform_totals.index,
    startangle=110,
    colors=["gray","#1aaa10"]
)

pylot.show()
```

```python
fig,ax= pylot.subplots(figsize=(8,8))


wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors=["gray", "#1aaa10"],)

ax.set_title(
    "Near Even Split between premium and free users",
    fontsize= 18,
    pad=20,
    loc="center"
)

for autotext in autotexts:
    autotext.set_fontsize(16)
    autotext.set_color("white")
    autotext.set_weight("bold")

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5,0.9),
    ncol=2
)


pylot.tight_layout()
pylot.show()

```

```python
pylot.figure(figsize=(10,6))

pylot.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=15,
    color="#2fba28",
    alpha=0.9,
    edgecolor="white"

)

pylot.xlabel("Avg Stream Duration (Minutes)")
pylot.ylabel("Number of Tracks")
pylot.title("Distribution of Average Stream Duration",fontsize=16,pad=20)

ax=pylot.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


pylot.show()
```

```python
numeric_spotify_df= spotify_df.select_dtypes(include=[np.number])
corr=numeric_spotify_df.corr()

im =pylot.imshow(corr, vmin=-1, vmax=1, cmap = "coolwarm")
pylot.colorbar(im, label="Pearson Corellation")

pylot.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
pylot.yticks(range(len(corr.columns)), corr.columns)

pylot.show()
```
